```python
q1 = kolom_numerik.quantile(0.25)
q3 = kolom_numerik.quantile(0.75)
iqr = q3 - q1
print(iqr)
```

```python
# handle warning 
import warnings 
warnings.filterwarnings('ignore')

# outlier filter
df_noid_align, iqr_new = kolom_numerik.align(iqr, axis=1, copy=False, join='outer')
outlier_filter = (kolom_numerik < q1 - 1.5 * iqr_new) | (kolom_numerik > q3 + 1.5 * iqr_new)
outlier_filter
```
